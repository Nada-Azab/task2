## Check missing values

```python
for col,null in enumerate(data.isnull().sum()):
  print(data.columns.values[col]," :",np.int(null/len(data)*100))
```

```python
# # feature eng
date_list=['created_at','first_milestone_at','last_milestone_at','updated_at','founded_at']


for date_col in date_list :
  # estimate missing dates
  # data[date_col].interpolate(method='linear', inplace=True)
  data[date_col]         = pd.to_datetime(data[date_col])
  data['year'+date_col]  = data[date_col].dt.year
  data['month'+date_col] = data[date_col].dt.month
  data['day'+date_col]   = data[date_col].dt.day
  data.drop(inplace =True,columns=date_col,axis=1)
```

```python
data.head()
```

```python
new_data=data[['country_code','category_code']]
```

```python
data[['country_code','status','category_code']].isnull().sum()
```

```python
data.drop(columns=['country_code','Unnamed: 0','category_code'],inplace=True)
```

```python
new_data=pd.get_dummies(new_data)
```

### final data
